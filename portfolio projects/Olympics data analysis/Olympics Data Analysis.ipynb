```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
# Loading the datasets
athletes = pd.read_csv('Data/athlete_events.csv') 
regions = pd.read_csv('Data/noc_regions.csv')
athletes.head(5)
```

```python
regions.head(5)
```

```python
# Join the two datasets on NOC
athlete_data = athletes.merge(regions, how= 'left', on = 'NOC')
athlete_data.head(5)
```

```python
# Removing unnecessary column notes from the dataset
athlete_data.drop(columns= 'notes',axis = 1, inplace = True)
athlete_data.head()
```

```python
athlete_data.shape
```

```python
# Renaming columns for consistency purposes
athlete_data.rename(columns = {'region':'Regions'}, inplace = True)
athlete_data.head()
```

```python
# Getting information about the data 
athlete_data.info()
```

```python
# Descriptive statistics about the dataset
athlete_data.describe()
```

```python
# Checking the data for missing values
nan_values = athlete_data.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# Sum of total nan values in the columns
athlete_data.isnull().sum()
```

```python
# Kenyas Olympics team details
athlete_data.query('Team=="Kenya"').head()
```

```python
# Kenya's Olympics gold medalists
kenyan_gold_medalists = athlete_data.loc[athlete_data.query('Team=="Kenya" & Medal=="Gold"').index]
kenyan_gold_medalists.head()
```

```python
# Kenya's Olympics Silver Medalists
kenyan_silver_medalists = athlete_data.loc[athlete_data.query('Team=="Kenya" & Medal=="Silver"').index]
kenyan_silver_medalists.head()
```

```python
# Kenya's Olympic Bronze Medalists
kenyan_silver_medalists = athlete_data.loc[athlete_data.query('Team=="Kenya" & Medal=="Bronze"').index]
kenyan_silver_medalists.head()
```

```python
# Top Countries participating in the competition.

top_10_countries = athlete_data.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
# Plot for the top 10 countries
plt.figure(figsize=(12,6))
plt.title('Olympic Participation Per Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette = 'Set2')
```

```python
# Age distribution of the athletes

plt.figure(figsize = (12,6))
plt.title('Age distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athlete_data.Age, bins= np.arange(10,80,2), edgecolor = 'white')
```

```python
# Olympics participation by gender
gender_counts = athlete_data.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize = (12,6))
plt.title('Olympics Gender Participation')
plt.pie(gender_counts, labels = gender_counts.index, autopct='%1.1f%%', startangle=150, shadow=True )
```

```python
# Total medals
medals = athlete_data.Medal.value_counts()
medals
```

```python
plt.figure(figsize = (12,6))
plt.title('Total Olympics Medal Share')
plt.pie(medals, labels = medals.index, autopct='%1.1f%%', startangle=150, shadow=True)
```

```python
# Female participation in the Olympics
female_participants = athlete_data.query('Sex=="F" & Season=="Summer"')
female_participants = female_participants[['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
women_participants = athlete_data.query('Sex=="F" & Season=="Summer"')
sns.set(style = 'darkgrid')
plt.figure(figsize = (20,10))
plt.title('Women Olympics Participation')
sns.countplot(x='Year', data=women_participants, palette='Spectral')
```
